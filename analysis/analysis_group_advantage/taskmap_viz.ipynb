```python
import pandas as pd
import numpy as np
import pickle
import ast
from tqdm import tqdm
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
from statsmodels.stats.multitest import fdrcorrection
import re
import itertools
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import ElasticNetCV, ElasticNet
from sklearn.metrics import r2_score, mean_squared_error
from sklearn.utils import resample
from sklearn.model_selection import train_test_split
import warnings
from sklearn.exceptions import ConvergenceWarning # yes, this is a thing; we have to ignore because it clutters the output
from scipy.stats import mannwhitneyu, pearsonr, chi2_contingency
from matplotlib.patches import Patch
import plotly.graph_objects as go
import plotly.express as px
from sklearn.decomposition import PCA

plt.rcParams["font.family"] = "Futura"

df_task_map = pd.read_csv('../data/task_map.csv', index_col='task')
```

```python
# Perform PCA
# scale the features
features_scaled = df_task_map.copy()
scaler = StandardScaler()
features_scaled = scaler.fit_transform(features_scaled)

pca = PCA(n_components=2)
pca_result = pca.fit_transform(features_scaled)

# Create a DataFrame with PCA results
pca_df = pd.DataFrame(data=pca_result, columns=['PC1', 'PC2'])
pca_df.index = df_task_map.index

# Plot PCA scatterplot with binary DV coloring
plt.figure(figsize=(10, 8))
scatter = plt.scatter(
    x=pca_df['PC1'], y=pca_df['PC2'],
    c = "gray",
    alpha=0.8
)
#plt.scatter(x=pca_df.loc['Public goods game', 'PC1'], y=pca_df.loc['Public goods game', 'PC2'], color='#EB8A90', s=100, alpha=1.0)
plt.scatter(
    x=[pca_df.loc['Public goods game', 'PC1']]*10, 
    y=[pca_df.loc['Public goods game', 'PC2']]*10, 
    color='#EB8A90', 
    s=[40000, 36000, 32000, 28000, 24000, 20000, 16000, 12000, 8000, 4000],  # Increase these values for a larger effect
    alpha=0.1, 
    zorder=0
)
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.show()
```
